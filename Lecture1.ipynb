```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
data = pd.read_csv("/content/lecture1/Salary Data (2).csv")
print(data)
print(data.columns)
```

```python
x = data["Years of Experience"]
y = data["Salary"]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x = data["Education Level"]
y = data["Salary"]

edu_levels = x.unique()
edut_to_int = {edu_levels[i]: i for i in range(len(edu_levels))}
print(edut_to_int)

x = x.map(edut_to_int)
print(x)
```

```python
fig, ax = plt.subplots()

ax.set_xlabel("Education Level")
ax.set_ylabel("Salary")



ax.bar(x,y)
```

```python
x = data["Job Title"]


x = x.value_counts()

x = x.head(10)

plt.pie(x, labels=x.index,autopct="%1.1f%%", startangle=140)

plt.title('Top 10 stanowisk pod wzgldem liczebnoci w firmie')
plt.axis('equal')
plt.show()
```
